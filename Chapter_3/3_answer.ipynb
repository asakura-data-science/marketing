### (3) 商品ごとに集計したい場合に利用されるpandasの命令はなんでしたか？

### (5) 公開データには4つの業態（コンビニエンスストア，スーパーマーケット，ホームセンター，薬粧店・ドラッグストア）のデータがある．集計分析を通じて業態の特徴について議論しよう． ../Chapter_5/in/datQpr.csvを利用し、例えば，業態名」と購入された「曜日」もしくは「中分類名」で購入数量を集計し，その構成比率から違いを見つけ，その背後の状況について考えてみよう．

```python
df = pd.read_csv('../Chapter_5/in/datQpr.csv', parse_dates=['日付'])

# 曜日を利用する場合
df['曜日'] = df['日付'].dt.dayofweek
df_qtty = df[['業態名', '曜日', '購入数量']].groupby(['業態名', '曜日']).sum()
df_qtty['ratio'] = df_qtty[['購入数量']].groupby(
    ['業態名']).apply(lambda x: x / x.sum())

display(df_qtty)

# 中分類名を利用する場合
df_qtty = df[['業態名', '中分類名', '購入数量']].groupby(['業態名', '中分類名']).sum()
df_qtty['ratio'] = df_qtty[['購入数量']].groupby(
    ['業態名']).apply(lambda x: x / x.sum())
display(df_qtty)
```
