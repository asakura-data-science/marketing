```python
# コード 3.1 ライブラリの import と焼肉データの読み込み
import pandas as pd

# pandasのデータフレームとしてcsvファイルを読み込む
df = pd.read_csv('in/yakiniku_2018.csv', parse_dates=['日付'])
df.head()
```

# データの基本的な値を把握

```python
# コード3.2 焼肉データの基礎集計
# データの期間を確認してみよう
print(df['日付'].min())
print(df['日付'].max())

# 営業日数の確認
print(len(df['日付'].unique()))
```

```python
# 売上金額の合計を新しい列としてオリジナルデータdfに追加しておく(後のコードで利用するため)。
df['売上金額'] = df['数量'] * df['単価']
total = df['売上金額'].sum()
print(total)

# 売上金額トップ10の商品
top10 = df[['商品名', '売上金額']].groupby('商品名').sum()
top10 = top10.sort_values('売上金額', ascending=False).head(10)
top10
```

```python
# コード3.4 商品種類数の計算
print(len(df['商品名'].unique()))
```

```python
# コード 3.5 売上数量上位 10 商品
num10 = df[['商品名', '数量']].groupby('商品名').sum()
num10 = num10.sort_values('数量', ascending=False).head(10)
num10
```

# 月別の売上金額の比較 (折れ線グラフの描画)

```python
# japanize_matplotlibaがインストールされていない場合は以下を実行
#!pip install japanize_matplotlib

# コード 3.6 月別の売上金額を計算し可視化
import japanize_matplotlib
import os
os.makedirs('out', exist_ok=True)

# 年月を新しい列としてオリジナルデータdfに追加しておく(後のセルで利用するため)。
df['年月'] = df['日付'].dt.to_period('M')

# 月別の売上金額を計算
month_amt = df[['年月', '売上金額']].groupby(['年月']).sum()
# 折れ線グラフを出力
ax = month_amt.plot()
ax.get_figure().savefig('out/3-1a.png')  # ファイルにグラフを保存
```

```python
# コード 3.7 月別に一日あたりの平均金額を計算し可視化
ds = pd.DataFrame()
day_amt = df[['日付', '売上金額']].groupby('日付', as_index=False).sum()
day_amt['年月'] = day_amt['日付'].dt.to_period('M')
day_avg = day_amt[['年月', '売上金額']].groupby(['年月']).mean()
display(day_avg.head(12))

ax = day_avg.plot()
ax.get_figure().savefig('out/3-1b.png')
```

```python
# コード3.7は、multi indexを使った方法でも記述可能
ds = df[['年月', '日付', '売上金額']].set_index(['年月', '日付'])
ds = ds.groupby(['年月', '日付']).sum()
ds = ds.reset_index('日付', drop=True)
ds = ds.groupby('年月').mean()
display(day_avg.head(12))

ax = day_avg.plot()
ax.get_figure().savefig('out/3-1b.png')
```
