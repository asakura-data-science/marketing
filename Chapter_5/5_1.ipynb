```python
import pandas as pd
import nysol.take as nt
import os
os.makedirs('out', exist_ok=True)
```

```python
# コード 5.1 モニタ-商品の 2 部グラフを作成するコード
df = pd.read_csv('in/datQpr.csv')  # QPRのデータを読み込む
```

```python
df = df[df['業態名'] == 'スーパー']  # スーパーのみ選択
# 日本茶・麦茶ドリンクを買っているモニタ一覧を作る
sel = df[df['細分類名'] == '日本茶・麦茶ドリンク']['モニタ'].drop_duplicates()
# そのモニタのデータのみを選択する
df = df[df['モニタ'].isin(sel)]
# 全員が日本茶・麦茶ドリンクを購入しており、
# そのような商品を加えても意味がないので削除しておく
bigraph = df[df['細分類名'] != '日本茶・麦茶ドリンク']
# モニタ項目と商品項目とを選択し、重複を削除すれば２部グラフの出来上がり
bigraph = df[['モニタ', '商品']].drop_duplicates()
# CSVに保存しておく
bigraph.to_csv('out/bigraph.csv', index=False)

print(bigraph)
print('モニタ数:', len(sel))
print('商品数:', len(bigraph['商品'].drop_duplicates()))
print('2部グラフ全行数:', len(bigraph))
```

```python
# コード 5.2 極大 2 部クリークを列挙するコード
# ２〜３分かかる
nt.mbiclique(ei="out/bigraph.csv", ef="モニタ,商品", o="out/biclique.csv").run()
# 極大2部クリークの結果を閲覧する
biclique = pd.read_csv('out/biclique.csv')
biclique
```

```python
# コード 5.3 モニタのサイズが 10 以上、商品のサイズが 3 以上のバイクラスタを選択する
# l=にモニタ集合と商品集合の最小サイズを指定することが可能
nt.mbiclique(ei='out/bigraph.csv', ef='モニタ,商品',
             l='10,3', o='out/biclique_10_3.csv').run()
# 極大2部クリークの結果を閲覧する
biclique = pd.read_csv('out/biclique_10_3.csv')
biclique
```

```python
# 例題5.12の解答

df = pd.read_csv('in/datQpr.csv')  # QPRのデータを読み込む
df = df[df['業態名'] == 'スーパー']  # スーパーのみ選択
# 日本茶・麦茶ドリンクを買っているモニタ一覧を作る
sel = df[df['細分類名'] == '日本茶・麦茶ドリンク']['モニタ'].drop_duplicates()

# そのモニタのデータのみを選択する
df = df[df['モニタ'].isin(sel)]
# 全員が日本茶・麦茶ドリンクを購入しており、
# そのような商品を加えても意味がないので削除しておく
df = df[df['細分類名'] != '日本茶・麦茶ドリンク']
df['レシート'] = df['モニタ'].str.cat(df['日付'].astype(str))
bigraph = df[['レシート', '細分類名']].drop_duplicates()
bigraph.to_csv('out/bigraph2.csv', index=False)

print(bigraph)
print('モニタ数:', len(sel))
print('商品数:', len(bigraph['細分類名'].drop_duplicates()))
print('2部グラフ全行数:', len(bigraph))

# l=にモニタ集合と商品集合の最小サイズを指定することが可能
nt.mbiclique(ei='out/bigraph2.csv', ef='レシート,細分類名',
             l='50,3', o='out/biclique2.csv').run()
biclique2 = pd.read_csv('out/biclique2.csv')
biclique2
```
