### (1) 相関ルールを列挙する際にいくつかの指標があるが，それらのしきい値を変更して実行すると結果がどのように変わるか確かめてみよう．具体的にはsupport値とjaccard係数の値を変更してみよう．入力ファイルはout/conv.csv or out/drug.csvを利用すればよい



```python
import pandas as pd
import nysol.take as nt


# support値を0.1に変更して実行
nt.mtra2gc(tid='モニタ', item='細分類名', s='0.1', i='out/conv.csv', eo='out/prac1.csv').run(msg="on")
df=pd.read_csv("out/prac1.csv")
display(df)

# ドラッグストアを対象にsupport値を0.01でJaccard係数を0.3で実行（閾値を変えて実行してみよう）
nt.mtra2gc(tid='モニタ', item='細分類名', s='0.01', sim='J', th='0.3', i='out/drug.csv', eo='out/prac2.csv').run(msg="on")
df=pd.read_csv("out/prac2.csv")
display(df)

# コンビニエンスストアを対象にsupport値を0.01でJaccard係数を0.3で実行（閾値を変えて実行してみよう）
nt.mtra2gc(tid='モニタ', item='細分類名', s='0.01', sim='J', th='0.3', i='out/conv.csv', eo='out/prac3.csv').run(msg="on")
df=pd.read_csv("out/prac3.csv")
df
```

### (2) しきい値を変更して得られた結果をネットワークで視覚化して，コンビニエンスストアの特徴とドラッグストアの特徴を解釈してみよう．

```python
import networkx as nx
import matplotlib.pyplot as plt
import pathlib

def graphDraw(file):

     # 辺の情報を保存
    rsl = pd.read_csv(file)
    edge = rsl[['node1%0', 'node2%1']]
    edge.to_csv('out/edge.csv', index=False, header=False)


    # グラフの作成
    G = nx.Graph()

    # ファイルの読み込み
    G = nx.read_edgelist('out/edge.csv', delimiter=',', nodetype=str)
    print(nx.number_of_nodes(G))  # ノード数をカウントして出力
    print(nx.number_of_edges(G))  # エッジ数をカウントして出力

    # グラフ描画のレイアウトを決定。バネモデルでk=はノードの距離を決めるパラメーター
    pos = nx.spring_layout(G, k=0.6, seed=4)

    plt.figure(figsize=(15, 12))  # 図のサイズを決定 (単位はインチ)
    plt.subplots_adjust(left=0.1, right=1)
    plt.axis('off')  # 軸は描画しない
    # ノードは赤色 (r)で透過率 (alpha=)60%、エッジはシアン (c)、文字サイズ 15インチ,フォントYuGothic(Mac),MS Gothic(Win)
    nx.draw_networkx(G, pos, node_color='r', alpha=0.6,
                     edge_color='c', font_size=11, font_family='YuGothic')


graphDraw("out/prac2.csv")
graphDraw("out/prac3.csv")
```
